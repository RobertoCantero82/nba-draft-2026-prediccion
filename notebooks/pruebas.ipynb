```python
from nba_api.stats.endpoints import draftcombinestats
import pandas as pd
import time

temporadas = [str(y) for y in range(2000, 2027)]
frames = []

for t in temporadas:
    try:
        df = draftcombinestats.DraftCombineStats(season_all_time=t).get_data_frames()[0]
        frames.append(df)
        time.sleep(1)
    except:
        pass

combine_full = pd.concat(frames, ignore_index=True)
combine_full.to_csv('combine_historico.csv', index=False)
print(combine_full.shape)
```

```python
# Buscar por nombre directamente
nombres = ['Mara', 'Miller', 'De Larrea']
combine_full[combine_full['LAST_NAME'].isin(nombres)][['PLAYER_NAME', 'SEASON', 'POSITION']]
```

```python
# Rellenamos los NaN de De Larrea con sus medidas conocidas
# Convertimos las columnas numéricas al tipo correcto
combine_full['HEIGHT_WO_SHOES'] = pd.to_numeric(combine_full['HEIGHT_WO_SHOES'], errors='coerce')
combine_full['WEIGHT'] = pd.to_numeric(combine_full['WEIGHT'], errors='coerce')
combine_full['WINGSPAN'] = pd.to_numeric(combine_full['WINGSPAN'], errors='coerce')

# Ahora sí rellenamos
mask = combine_full['LAST_NAME'] == 'De Larrea'
combine_full.loc[mask, 'HEIGHT_WO_SHOES'] = 77.95
combine_full.loc[mask, 'WEIGHT'] = 175.0
combine_full.loc[mask, 'WINGSPAN'] = 79.5
```
